```python
def price_dep(col, df):
    _2019 = df[df['year']==2019]
    _2017 = df[df['year']==2017]
    _2015 = df[df['year']==2015]
    _2013 = df[df['year']==2013]
    _2019 = _2019[_2019['price']<200000]
    _2017 = _2017[_2017['price']<200000]
    _2015 = _2015[_2015['price']<200000]
    _2013 = _2013[_2013['price']<200000]
    _2019 = _2019.groupby(col).mean()[['price']].reset_index()
    _2017 = _2017.groupby(col).mean()[['price']].reset_index()
    _2015 = _2015.groupby(col).mean()[['price']].reset_index()
    _2013 = _2013.groupby(col).mean()[['price']].reset_index()
    merged = _2019.merge(_2017,on=col,how='inner').merge(_2015,on=col,how='inner').merge(_2013,on=col,how='inner')
    merged.columns = ['Target','2019','2017','2015','2013']
    merged['Average Difference'] = [((a-b)+(b-c)+(c-d))/3 for a, b, c, d in zip(list(merged['2019']), list(merged['2017']), list(merged['2015']), list(merged['2013']))]
    return merged
    
```

```python
def find_states(df):
    x = df['city'].apply(lambda x: x[-2:] if x[-2:].isupper() else float('NaN')).dropna()
    weight = []
    for _ in x.groupby(x).count():
        weight += [round(_/sum(x.groupby(x).count()),4)]
    used_car_weights = pd.DataFrame({
    'State':list(x.groupby(x).count().index),
    'Weight':weight
    }).sort_values('Weight', ascending=False).reset_index().drop('index',axis=1)
    return used_car_weights
```
